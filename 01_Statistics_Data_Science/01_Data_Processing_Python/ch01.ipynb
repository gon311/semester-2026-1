```python
a = 0
b = 2.5
c = True
d = "안녕하세요"
e = [1, 2.5, "hello"]  # list : order O
f = {"name" : "han"} # dictionary : key-value
g = (1, 2, 3) # tuple :  order O,  Immutable
h = {1, 2, 2, 3} # set : order X, duplicate X

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(e[2])
print(f["name"]) # key indexing
print(g[1])
# print(h[0]) => 'set' object is not subscriptable => 값을 얻으려면 for문을 사용해 전체를 순회하거나, list() 또는 tuple()로 변환.
```

## 자료 인덱싱, 슬라이싱

```python
# indexing & slicing
text = "Hello World"
print(text[0]) # indexing
print(text[0:5])
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])
```

## 조건문
- if
- if ~ elif
- if ~ elif ~ else

```python
# conditional
# if/elif/else

x = 10
if x > 0 :
    print("x는 양수입니다.")
elif x == 0:
    print("x는 0입니다.")
else:
    print("x는 음수입니다.")
```

## 반복문
- for
- while

```python
# for
nums = [0, 1, 2, 3, 4]

for i in nums:
    print(i)

```

```python
count = 10
while count >= 0:
    print(count)
    count = count - 1

print("발사!")
```

## 함수 정의

```python
# 함수의 정의와 호출

def add(a, b):
    return a + b

print(add(2, 4))



def greet(name):
    print(f" 안녕하세요. {name}님!")

greet("홍길동")
```
